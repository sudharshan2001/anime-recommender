```python
synopsis1=[]
for i in url_list:
    try:
        r = requests.get(i)
        soup = BeautifulSoup(r.content,'html5lib')
        synopsis1.append(soup.find_all('p',attrs={'itemprop':"description"})[0].contents[0])
    except NameError:
        synopsis1.append('No synopsis information has been added to this title. Help improve our database by adding a synopsis')
    except IndexError:
        synopsis1.append('No synopsis information has been added to this title. Help improve our database by adding a synopsis')
```

```python
data_syn=pd.DataFrame({'uid_list':uid_list,'title_list':title_list,'synopsis':synopsis1})
data_syn
```

```python
for i,row in fo_synopsis.iterrows():
    if row['uid'] in uid_list: 
        fo_synopsis.at[i,'synopsis'] = data_syn[data_syn['uid_list']==row['uid']]['synopsis'].values[0]
```

```python
epi_url_list=[]
epi_uid_list=[]

for i in fo_synopsis[pd.isnull(fo_synopsis['episodes'])].iterrows():
    epi_uid_list.append(list(i)[1][0])
    epi_url_list.append(list(i)[1][-1])
```

```python
epi_list=[]
for link in epi_url_list:
    try:
        r = requests.get(link)
        soup = BeautifulSoup(r.content,'html5lib')
        episode=float(soup.find_all('div', attrs={'class':'user-status-block js-user-status-block fn-grey6 clearfix al mt8 po-r'})[0].find_all('span')[0].contents[0])
        epi_list.append(episode)
    except (ValueError,IndexError):
        epi_list.append('N/A')
```

```python
fo_synopsis['episodes']=fo_synopsis['episodes'].astype(object)
```

```python
data_epi=pd.DataFrame({'uid_list':epi_uid_list,'episode':epi_list})
data_epi
```

```python
for i,row in fo_synopsis.iterrows():
    if row['uid'] in epi_uid_list: 
        fo_synopsis.at[i,'episodes'] = data_epi[data_epi['uid_list']==row['uid']]['episode'].values[0]
```

```python
rank_uid_list=[]
rank_url_list=[]

for i in fo_synopsis[pd.isnull(fo_synopsis['ranked'])].iterrows():
    rank_uid_list.append(list(i)[1][0])
    rank_url_list.append(list(i)[1][-1])
```

```python
fo_synopsis['ranked']=fo_synopsis['ranked'].astype(object)
```

```python
rank_list=[]
for link in rank_url_list:
    try:
        r = requests.get(link)
        soup = BeautifulSoup(r.content,'html5lib')
        ranked = soup.find_all('span', attrs={'class':'numbers ranked'})[0].find_all('strong')[0].contents[0]
        if '#' in ranked:
            rank_list.append(ranked[1:])
        else:
            rank_list.append(ranked)
    except (ValueError,IndexError):
        rank_list.append('N/A')
```

```python
data_rank=pd.DataFrame({'uid_list':rank_uid_list,'rank':rank_list})
data_rank.head(1)
```

```python
for i,row in fo_synopsis.iterrows():
    if row['uid'] in rank_uid_list: 
        fo_synopsis.at[i,'ranked'] = data_rank[data_rank['uid_list']==row['uid']]['rank'].values[0]
```

```python
score_uid_list=[]
score_url_list=[]

for i in fo_synopsis[pd.isnull(fo_synopsis['score'])].iterrows():
    score_uid_list.append(list(i)[1][0])
    score_url_list.append(list(i)[1][-1])
```

```python
score_list=[]

for link in score_url_list:
    try:
        r = requests.get(link)
        soup = BeautifulSoup(r.content,'html5lib')
        score=soup.find_all('div', attrs={'class':'fl-l score'})[0].find_all('div')[0].contents[0]
        score_list.append(score)
    except (ValueError,IndexError):
        score_list.append('N/A')
```

```python
data_score=pd.DataFrame({'uid_list':score_uid_list,'score':score_list})
data_score.tail(15)
```

```python
fo_synopsis['score']=fo_synopsis['score'].astype(object)
```

```python
for i,row in fo_synopsis.iterrows():
    if row['uid'] in score_uid_list: 
        fo_synopsis.at[i,'score'] = data_score[data_score['uid_list']==row['uid']]['score'].values[0]
```
